```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
```

```python
tau = 11.9245
A = 21.908
B = 2.5
tau_vals = []
```

```python
def func(time, tau):
    return A*B*(1-np.exp(-time/tau))
```

```python
def manual_solve(time_vals, velocity_vals):
    #Manual solve
    tau_vals = []
    for i in range(time_vals.size):
        #solve for tau
        new_tau = -time_vals[i]/(np.log(1-velocity_vals[i]/(A*B)))
        tau_vals.append(new_tau)
    return np.nanmean(tau_vals)
```

```python
df = pd.read_csv("1.csv")
time_vals = df['Time(s)']
velocity_vals = df['Velocity']

#remove first 60 rows of deadtime (by inspection)
time_vals = time_vals.iloc[60:]
velocity_vals=velocity_vals.iloc[60:]

#reset indexes
time_vals = time_vals.reset_index(drop=True)
velocity_vals = velocity_vals.reset_index(drop=True)

#shift to left
initial_value = time_vals[0]
time_vals = time_vals-initial_value
plt.plot(time_vals, velocity_vals)
```

```python
popt, pcov = curve_fit(func, time_vals, velocity_vals, p0=tau)
tau_vals.append(*popt)
popt
```

```python
manual_solve(time_vals, velocity_vals)
```

```python
f, axarr = plt.subplots(3, sharex = True)
f.tight_layout()
axarr[0].plot(time_vals, velocity_vals)
axarr[0].set_title("Actual vals")

axarr[1].plot(time_vals, func(time_vals, tau))
axarr[1].set_title("Initial Tau")

axarr[2].plot(time_vals, func(time_vals, *popt))
axarr[2].set_title("Best fit(?)")

plt.show()
```
